```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
X_train, X_test, y_train, y_test=train_test_split(cancer['data'],cancer['target'],stratify=cancer['target'],random_state=0)
```

```python
train_accuracy=[]
test_accuracy=[]
num_neighbors=range(1,11)
```

```python
import matplotlib.pyplot as plt
for n_neighbors in num_neighbors:
  cls=KNeighborsClassifier(n_neighbors=n_neighbors)
  cls.fit(X_train,y_train)
  train_accuracy.append(cls.score(X_train,y_train))
  test_accuracy.append(cls.score(X_test,y_test))

plt.plot(num_neighbors,train_accuracy,label='train accuracy')
plt.plot(num_neighbors,test_accuracy,label='test accuracy')
plt.ylabel('accuracy')
plt.xlabel('n-neighbors')
plt.legend()
```
